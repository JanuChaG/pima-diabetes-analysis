```python
import pandas as pd

# Load dataset
df = pd.read_csv(r"C:\Users\Jayan Gamage\pima-diabetes-analysis\data\diabetes.csv")

# Quick look at the data
print("First 5 rows:")
print(df.head())
print("\nDataset info:")
print(df.info())

# Replace zeros in key columns with NaN
cols_to_fix = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
df[cols_to_fix] = df[cols_to_fix].replace(0, pd.NA)

# Fill missing values with column mean
df = df.fillna(df.mean())

# Save cleaned dataset
df.to_csv(r"C:\Users\Jayan Gamage\pima-diabetes-analysis\data\cleaned_diabetes.csv", index=False)

print("\nCleaned dataset saved as cleaned_diabetes.csv")
```
